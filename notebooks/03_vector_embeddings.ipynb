```python
import pandas as pd
# instal sentence-transformers (pip install sentence-transformers)
from sentence_transformers import SentenceTransformer
import numpy as np

# Load Data
df = pd.read_csv("/content/cleaned_dataset.csv")
deskripsi_produk = df['Deskripsi Teks (Untuk Embeddings)'].tolist()

# Gunakan pre-trained model bahasa natural (Multilingual/Indonesia)
print("Memuat model embedding...")
embedding_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Proses transformasi teks ke matriks vektor padat (Dense Vectors)
print("Menghasilkan koordinat vektor (Proses ini membutuhkan waktu komputasi)...")
product_vectors = embedding_model.encode(deskripsi_produk, show_progress_bar=True)

print(f"Dimensi matriks vektor yang dihasilkan: {product_vectors.shape}")
```

```python
import pickle
import os

# Simpan vektor ke dalam disk lokal (sebelum diunggah ke Vector DB)
vector_path = "../data/embeddings/product_vectors.pkl"

# Buat direktori
os.makedirs(os.path.dirname(vector_path), exist_ok=True)

with open(vector_path, "wb") as f:
    pickle.dump({'id': df['ID Produk'].tolist(), 'vectors': product_vectors}, f)

print(f"Vector embeddings berhasil disimpan ke {vector_path}")
```
